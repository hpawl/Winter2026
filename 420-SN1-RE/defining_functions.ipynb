### Function 3:
Create a function to calculate the average of three separate numbers

```python
# Input & Variables
x = int(input("Enter first number \"X\": "))
y = int(input("Enter second number \"Y\": "))

# Functions
def sum(x, y):
    result = x + y
    return result

def avg(x, y):
    result = sum(x, y) // 2
    return result

# Output & Calling
avg_of_numbers = avg(x, y)
print(f"The average of {x} and {y} is {avg_of_numbers}")
```

Now average of three numbers:

```python
# Input & Variables
x = float(input("Enter first number \"X\": "))
y = float(input("Enter second number \"Y\": "))
z = float(input("Enter third number \"Z\": "))

# Functions
def sum(x, y, z = 0):
    result = x + y + z
    return result

def avg(x, y, z = "NA"):
    xy_sum = sum(x, y)
    if z == "NA":
        result = xy_sum // 2
    else:
        result = sum(xy_sum, z) // 3
    return result

# Output & Calling
avg_of_numbers = avg(x, y, z)
print(f"The average of {x}, {y}, and {z} is {avg_of_numbers}")
```
